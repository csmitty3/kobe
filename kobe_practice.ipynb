```python
from matplotlib import pyplot as plt
file = "Kobe.csv"
ppg = []
season = []
count = 0
for line in open(file):
    line = line.replace("\n", "")
    line = line.split(",")
    if count > 0:
        ppg.append(float(line[29]))
        season.append(line[0])
    
    count += 1
ppg.pop(20)
season.pop(20)
print(ppg)
print(season)
plt.figure(figsize = (20, 10), dpi = 80)
plt.plot(season, ppg, marker = "*")
plt.show()
```
